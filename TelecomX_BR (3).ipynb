```python
import requests
import pandas as pd

url = 'https://raw.githubusercontent.com/alura-cursos/challenge2-data-science/refs/heads/main/TelecomX_Data.json'
response = requests.get(url)
data = response.json()
df = pd.DataFrame(data)

print(df.head())
```

#ðŸ”§ TransformaÃ§Ã£o

```python
from pandas import json_normalize

df = json_normalize(data)
df.head()
```

```python
customer_expanded = json_normalize(df['customer'])
```

```python
df = pd.concat([df.drop('customer', axis=1), customer_expanded], axis=1)
```

```python
phone_expanded = json_normalize(df['phone'])
df = pd.concat([df.drop('phone', axis=1), phone_expanded], axis=1)
```

```python
internet_expanded = json_normalize(df['internet'])
df = pd.concat([df.drop('internet', axis=1), internet_expanded], axis=1)
```

```python
account_expanded = json_normalize(df['account'])
df = pd.concat([df.drop('account', axis=1), account_expanded], axis=1)

```

```python
df['Charges.Monthly'] = pd.to_numeric(df['Charges.Monthly'], errors='coerce')
```

```python
num_cols = df.select_dtypes(include=['float64', 'int64'])
desc = num_cols.describe().T
desc['median'] = num_cols.median()
desc[['mean', 'median', 'std', 'min', '25%', '50%', '75%', 'max']]
```

```python
categorical_vars = ['gender', 'Contract', 'PaymentMethod']  # Ajuste conforme suas colunas

for var in categorical_vars:
    print(f"Valores Ãºnicos em {var}: {df[var].unique()}\n")
```

```python
for var in categorical_vars:
    churn_by_var = df.groupby([var, 'Churn']).size().unstack(fill_value=0)
    churn_by_var['Total'] = churn_by_var.sum(axis=1)
    churn_by_var['Churn_Rate_%'] = (churn_by_var.get('Yes', 0) / churn_by_var['Total']) * 100
    print(f"Churn por {var}:\n", churn_by_var, "\n")
```

```python
for var in categorical_vars:
    plt.figure(figsize=(8,4))
    sns.countplot(data=df, x=var, hue='Churn')
    plt.title(f'Churn por {var}')
    plt.xticks(rotation=45)
    plt.show()
```

```python
churn_by_gender = df.groupby(['gender', 'Churn']).size().unstack(fill_value=0)
churn_by_gender['Total'] = churn_by_gender.sum(axis=1)
churn_by_gender['Churn_Rate_%'] = (churn_by_gender.get('Yes', 0) / churn_by_gender['Total']) * 100
print(churn_by_gender)
```
